```python
from ultralytics import YOLO
import cv2

model = YOLO("best.pt")

video_path = "watermarked_preview.mp4"   

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(
        frame,
        conf=0.25,
        iou=0.5,
        tracker="bytetrack.yaml",
        persist=True
    )

    annotated_frame = results[0].plot()

    cv2.imshow("Construction Safety Monitoring", annotated_frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
